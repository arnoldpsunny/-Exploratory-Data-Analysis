```python
df['Year']=df['Year'].str.replace('–','')
df['Year']=df['Year'].str.replace('(','')
df['Year']=df['Year'].str.replace(')','')
```

```python
df.head()
```

```python
df['Year']=pd.to_numeric(df['Year'], errors = 'coerce')
```

```python
df['Minutes']=df['Minutes'].str.replace(' min','')
```

```python
df['Minutes']=pd.to_numeric(df['Minutes'], errors = 'coerce')
```

```python
df['Votes']=pd.to_numeric(df['Votes'], errors = 'coerce')
```

```python
df[['genre1','genre2','genre3']]=df['genre'].str.split(pat = ',', n=2,expand = True)
```

```python
col=['Year','Rating','Minutes',	'Votes']
for c in col:
  plt.figure()
  sns.boxplot(y=c, data=df)
```

```python
col=['Year','Rating','Minutes',	'Votes']
for c in col:
  plt.figure()
  sns.boxplot(y=c,data=df)
  percentile25=df[c].quantile(0.25)
  percentile75=df[c].quantile(0.75)
  iqr=percentile75-percentile25
  upper_limit=percentile75+(1.5*iqr)
  lower_limit=percentile25-(1.5*iqr)
  df=df[(df[c]<=upper_limit) & (df[c]>=lower_limit)]
  plt.figure()
  sns.boxplot(y=c,data=df)
```
