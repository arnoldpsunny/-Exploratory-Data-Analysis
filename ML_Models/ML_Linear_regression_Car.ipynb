```python
# Capping Winzonarisation
col=['Torque_Nm','Power_bhp','Fuel Tank Capacity','Seating Capacity','Width','Engine','Kilometer','Year','Price']
for c in col:
    lower_limit=df[c].quantile(0.1)
    upper_limit=df[c].quantile(0.9)
    df[c]=np.where(df[c]>= upper_limit,
            upper_limit,
            np.where(df[c]<=lower_limit,
                    lower_limit,
                    df[c]))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
X = df.drop(["Price"], axis=1)
Y = df["Price"]
```

```python
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.3, random_state=4)
```

```python
X_train=sc.fit_transform(X_train)
X_train
```

```python
X_test=sc.transform(X_test)
X_test
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
reg.fit(X_train, Y_train)
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
Y_pred=reg.predict(X_test)
```

```python
from sklearn import metrics
metrics.mean_squared_error(Y_test,Y_pred)
```

```python
np.sqrt(metrics.mean_squared_error(Y_test,Y_pred))
```

```python
metrics.r2_score(Y_test, Y_pred)
```
