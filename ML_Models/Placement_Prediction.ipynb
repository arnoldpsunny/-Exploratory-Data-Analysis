# Outlier Analysis

```python
df
```

```python
sns.boxplot(y="degree_p",data=df)
```

```python
percentile25=df["degree_p"].quantile(0.25)
```

```python
percentile75=df["degree_p"].quantile(0.75)
```

```python
iqr=percentile75-percentile25
```

```python
iqr
```

```python
upper_limit=percentile75+(1.5*iqr)
```

```python
lower_limit=percentile25-(1.5*iqr)
```

```python
upper_limit
```

```python
lower_limit
```

```python
df=df[(df["degree_p"]<=upper_limit) & (df["degree_p"]>=lower_limit)]
```

```python
sns.boxplot(y="degree_p",data=df)
```

```python
df.head(1)
```

```python
col=["ssc_p","hsc_p","etest_p","mba_p","salary"]
```

```python
for c in col:
  plt.figure()
  sns.boxplot(y=c, data=df)
```

```python
col=["hsc_p","salary"]
for c in col:
  plt.figure()
  sns.boxplot(y=c,data=df)
  percentile25=df[c].quantile(0.25)
  percentile75=df[c].quantile(0.75)
  iqr=percentile75-percentile25
  upper_limit=percentile75+(1.5*iqr)
  lower_limit=percentile25-(1.5*iqr)
  df=df[(df[c]<=upper_limit) & (df[c]>=lower_limit)]
  plt.figure()
  sns.boxplot(y=c,data=df)
```

# Clustering

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=6)
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
sc.fit_transform(df)
```

```python
scaled_df=pd.DataFrame(sc.fit_transform(df), columns=df.columns)
```

```python
scaled_df.head(1)
```

```python
kmeans.fit(scaled_df)
```

```python
kmeans.predict(scaled_df)
```

```python
scaled_df["clusters"]=kmeans.predict(scaled_df)
```

```python
scaled_df
```

```python
kmeans.inertia_
```

Elbow Method

```python
scaled_df.drop(["clusters"],axis=1, inplace=True)
```

```python
inertia_list=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i)
  kmeans.fit(scaled_df)
  inertia_list.append(kmeans.inertia_)
```

```python
inertia_list
```

```python
range(1,31)
```

```python
plt.plot(range(1,11), inertia_list, marker="o")
plt.xlabel("Number of clusters")
plt.ylabel("Inertia")
plt.title("Elbow curve")
plt.show()
```
